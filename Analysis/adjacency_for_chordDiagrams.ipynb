```python
data_path = r'../data/filtered_gtd_no_summary.csv'
data_df = pd.read_csv(data_path, delimiter=',',encoding='ISO-8859-1')
```

```python
# extract the 30 most common groups
c = Counter(data_df.gname)
groups = c.most_common()[1:31] # Unknown is the most common
gnames = [i[0] for i in groups]
gnames
```

```python
# only takes entries of these 30 groups
clean_df = data_df.loc[data_df.gname.apply(lambda x: x in gnames)]
c=Counter(clean_df.region_txt)
c.most_common()
```
